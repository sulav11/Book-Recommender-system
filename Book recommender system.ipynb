```python
import pandas as pd
```

```python
books = pd.read_csv('Books.csv')
ratings = pd.read_csv('Ratings.csv')
users = pd.read_csv('Users.csv')
```

```python
df1 = books.merge(ratings,how="left", on="ISBN")
df= df1.merge(users,how="left", on="User-ID")


df.head()
```

```python
df.drop(columns=["Image-URL-S","Image-URL-M",'Location','Age','Year-Of-Publication','Publisher'],inplace=True)
df
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna()
df
```

```python
df['User-ID'].value_counts()
```

```python
x = df['User-ID'].value_counts()>200
x
```

```python
y = x[x].index
y
```

```python
# we only take the users who have got valid knowledge of books .So we limit down the no of users to 811
df = df[df['User-ID'].isin(y)]
df
```

```python
# We are only taking books with 50+ users rating for its validity
x = df['Book-Title'].value_counts()>50
x
```

```python
a = x[x].index
a
```

```python
df = df[df['Book-Title'].isin(a)]
```

```python
df
```

```python
df['Book-Title'].value_counts()
```

```python
df['Total Votes'] = df.groupby(['Book-Title'])['Book-Rating'].transform('count')
```

```python
df
```

```python
ratings = df.groupby('Book-Title', as_index=False)['Book-Rating'].mean()
ratings['Book-Rating'] = ratings['Book-Rating'].apply(lambda x : round(x, 1))
ratings = ratings.rename(columns = {'Book-Rating': 'Avg_rating'})
```

```python
ratings
```

```python
df['User-ID'] = df['User-ID'].apply(lambda x : int(x))
df.merge(ratings,on = 'Book-Title')
```

```python
from scipy.sparse import csr_matrix
```

```python
df = df.drop_duplicates(['User-ID','Book-Title'])
df = df.reset_index()
df
```

```python
pivot_table = df.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
pivot_table.shape
```

```python
pivot_table.fillna(0 ,inplace = True)
```

```python
pivot_table
```

```python
book_sparse = csr_matrix(pivot_table)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric = 'cosine' ,algorithm='brute',n_neighbors = 6)
model.fit(book_sparse)
```

```python
import numpy as np
```

```python
def recommend(book):
    n = np.where(pivot_table.index == book)[0][0]
    model.fit(book_sparse)
    distances, suggestions = model.kneighbors(pivot_table.iloc[n, :].values.reshape(1, -1))
   
    for i in range(6):
        print(pivot_table.index[suggestions[0,i]])
```

```python

recommend('1984')
```
